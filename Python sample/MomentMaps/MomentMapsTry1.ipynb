0th moment (gas mass)
1st moment (velocity towards / away from us).
2nd moment tells you the temperature of the gas at that pixel.

```python

import spectral_cube
from spectral_cube import SpectralCube
import aplpy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from astropy.io import fits
from astropy.wcs import WCS
import matplotlib.ticker as tck
from astropy import units as u

#, 'L1641S_N2H+10'

cubefiles = ['L1641S_C18O10', 'L1641S_HCO+10', 'L1641S_HNC10', 'L1641S_N2H+10']


for i in cubefiles:    
    
    with fits.open('/home/asus/AnacondaProjects/Data/FitsCubes/%s.fits'%(i)) as ff:
        
        cube = SpectralCube.read(ff)
        
        print(cube)

        cube=cube*u.K #here we give units to out unitless cube

        timessigma=1
        
        ston = 4 #we perform a cut corresponding to the s/n implemented in the fitter of N2H+
        #s/n = 4 -> s > n*4
        
        if i == 'L1641S_C18O10':
            #mask = cube > (-6.8808084831884265e-06+timessigma*8.01791510975679e-05)*ston*u.K #surprisingly results change when I add this mask
            mask = cube > (-6.8808084831884265e-06)*u.K

        if i == 'L1641S_HCO+10':
            #mask = cube > (-1.7633809849137914e-06+timessigma*6.0880847112321756e-05)*ston*u.K #surprisingly results change when I add this mask
            mask = cube > (-1.7633809849137914e-06)*u.K

        if i == 'L1641S_HNC10':
            #mask = cube > (-1.3174341160825511e-05+timessigma*4.500864425241393e-05)*ston*u.K #surprisingly results change when I add this mask
            mask = cube > (-1.3174341160825511e-05)*u.K
        if i == 'L1641S_N2H+10':
            #mask = cube > (-2.251887573688305e-05+timessigma*7.173484576944488e-05)*ston*u.K #surprisingly results change when I add this mask
            mask = cube > (-2.251887573688305e-05)*u.K

        maskedcube=cube.with_mask(mask)
        #maskedcube=cube # uncomment when you dont want to add any mask

    #-----------------------------------------------------------------------------------------

    #lets compute first the linewidth map

        sigma_map = cube.with_spectral_unit(u.km/u.s).linewidth_sigma()
        hduprimary=sigma_map.hdu
        data=hduprimary.data
        #data=np.nan_to_num(data)

        print(sigma_map.wcs)

        fig = plt.figure()
        ax = fig.add_subplot(111,projection = sigma_map.wcs)
        #title configurations


        if i=='L1641S_C18O10':
            plt.title('$Linewidth \quad \sigma \quad map \quad for \quad C^{18}O (1-0)$')
            im = ax.imshow(data,origin='lower',vmax=8)

        elif i=='L1641S_HCO+10':

            plt.title('$Linewidth \quad \sigma \quad map \quad for \quad HCO^{+} (1-0)$')
            im = ax.imshow(data,origin='lower',vmax=10)

        elif i=='L1641S_HNC10':

            plt.title('$Linewidth \quad \sigma \quad map \quad for \quad HNC (1-0)$')
            im = ax.imshow(data,origin='lower',vmax=35)

        elif i=='L1641S_N2H+10':
            plt.title('$Linewidth \quad \sigma \quad map \quad for \quad N_{2}H^{+} (1-0)$')
            im = ax.imshow(data,origin='lower',vmax=9.5)

        cbar = plt.colorbar(im)
        cbar.set_label('%s'%sigma_map.unit)

        plt.xlabel('Right Ascension (J2000)')
        plt.ylabel('Declination (J2000)')


        print('linewidth map has units of: ',sigma_map.unit)

        plt.savefig('linewidth_sigma/%s_linewidth_sigma.png'%i,dpi=700, bbox_inches = 'tight')
        
        plt.savefig('/home/asus/Documents/Latex/Master_Thesis3/figures/%s_linewidth_sigma.pdf'%i,dpi=700,format='pdf', bbox_inches = 'tight')

    #--------------------------------------------------------------------------------------------

        fwhm_map = cube.with_spectral_unit(u.km/u.s).linewidth_fwhm()

        hduprimary=fwhm_map.hdu
        data=hduprimary.data
        #data=np.nan_to_num(data)

        print(fwhm_map.wcs)

        fig = plt.figure()
        ax = fig.add_subplot(111,projection = fwhm_map.wcs)
        #title configurations


        if i=='L1641S_C18O10':
            plt.title('$Linewidth \quad FWHM \quad map \quad for \quad C^{18}O (1-0)$')
            im = ax.imshow(data,origin='lower', vmax=16)

        elif i=='L1641S_HCO+10':
            plt.title('$Linewidth \quad FWHM \quad map \quad for \quad HCO^{+} (1-0)$')
            im = ax.imshow(data,origin='lower', vmax=24)

        elif i=='L1641S_HNC10':
            plt.title('$Linewidth \quad FWHM \quad map \quad for \quad HNC (1-0)$')
            im = ax.imshow(data,origin='lower', vmax=90)

        elif i=='L1641S_N2H+10':
            plt.title('$Linewidth \quad FWHM \quad map \quad for \quad N_{2}H^{+} (1-0)$')
            im = ax.imshow(data,origin='lower', vmax=20)

        print('linewidth map has units of: ',fwhm_map.unit)

        cbar = plt.colorbar(im)
        cbar.set_label('%s'%fwhm_map.unit)

        plt.xlabel('Right Ascension (J2000)')
        plt.ylabel('Declination (J2000)')

        plt.savefig('linewidth_FWHM/%s_linewidthFWHM.png'%i,dpi=700, bbox_inches = 'tight')
        plt.savefig('/home/asus/Documents/Latex/Master_Thesis3/figures/%s_linewidthFWHM.pdf'%i,dpi=700,format='pdf', bbox_inches = 'tight')

#--------------------------------------------------------------------------------------------    

        for j in (0,1,2):

            momentmap=maskedcube.with_spectral_unit(u.km/u.s).moment(order=j)
            momentmap.write('/home/asus/AnacondaProjects/Data/FitsCubes/moment%s/%s_moment_%s.fits'%(j,i,j), overwrite=True)
            hduprimary=momentmap.hdu
            data=hduprimary.data
            #data=np.nan_to_num(data)

            print(momentmap.wcs)

            fig = plt.figure()
            ax = fig.add_subplot(111,projection = momentmap.wcs)

            #lets set the configurations for moment=0 maps

            if j==0 and (i=='L1641S_HCO+10' or 'L1641S_HNC10' or 'L1641S_C18O10'):
                    im = ax.imshow(data,origin='lower',cmap='gist_heat', vmin=0
                               ,vmax=8)
                #to obtain column density I should multiply here like: moment_0 * 1.82 * 10**18 / (u.cm * u.cm) * u.s / u.K / u.km
            elif j==0 and i=='L1641S_N2H+10':
                im = ax.imshow(data,origin='lower',cmap='gist_heat', vmin=8)


            #lets set the configurations for moment=1 maps


            #range1=10
            #range2=7
            range3=24
            #range4=8

            if j==1 and i=='L1641S_C18O10':
                im = ax.imshow(data,origin='lower',cmap='seismic'
                              #,vmin=3.6,vmax=5.6
                              # , vmin=0,vmax=8
                              )
            elif j==1 and i=='L1641S_HCO+10':
                im = ax.imshow(data,origin='lower',cmap='seismic'
                              #,vmin=1.5,vmax=7
                              # , vmin=-2,vmax=8
                              )
            elif j==1 and i=='L1641S_HNC10':
                im = ax.imshow(data,origin='lower',cmap='seismic'
                              #, vmin=2.5,vmax=7
                              # , vmin=-range3,vmax=range3
                              )
            elif j==1 and i=='L1641S_N2H+10':
                im = ax.imshow(data,origin='lower',cmap='seismic'
                              # , vmin=0,vmax=9
                              # , vmin=-4,vmax=8
                              )

            #lets set the configurations for moment=2 maps

            range1=35
            range2=50
            #range3=50
            range4=30 #before 25

            if j==2 and i=='L1641S_C18O10':
                im = ax.imshow(data,origin='lower'
                              #, vmin=-range1
                               ,vmax=350
                              )
            elif j==2 and i=='L1641S_HCO+10':
                im = ax.imshow(data,origin='lower'
                              # , vmin=-range2
                               ,vmax=650
                              )
            elif j==2 and i=='L1641S_HNC10':
                im = ax.imshow(data,origin='lower'
                               , vmax=800, 
                               #,vmin=-5000
                              )
            elif j==2 and i=='L1641S_N2H+10':
                im = ax.imshow(data,origin='lower'
                              # , vmin=-range4,vmax=range4
                              )


            cbar = plt.colorbar(im)
            cbar.set_label('%s'%momentmap.unit)

            plt.xlabel('Right Ascension (J2000)')
            plt.ylabel('Declination (J2000)')

            #title configurations

            if i=='L1641S_C18O10':
                plt.title('$Moment=%s \quad map \quad for \quad C^{18}O (1-0)$'%j)

            elif i=='L1641S_HCO+10':

                plt.title('$Moment=%s \quad map \quad for \quad HCO^{+} (1-0)$'%j)

            elif i=='L1641S_HNC10':

                plt.title('$Moment=%s \quad map \quad for \quad HNC (1-0)$'%j)

            elif i=='L1641S_N2H+10':

                plt.title('$Moment=%s \quad map \quad for \quad N_{2}H^{+} (1-0)$'%j)


            print('Momentmap %s has units of: '%j,momentmap.unit)

            plt.savefig('moment%s_%ssigma/%s_moment%s.png'%(j,timessigma,i,j),dpi=700, bbox_inches = 'tight')
            plt.savefig('/home/asus/Documents/Latex/Master_Thesis3/figures/moment%s_%ssigma_%s_moment%s.pdf'%(j,timessigma,i,j),dpi=700,format='pdf', bbox_inches = 'tight')

```
